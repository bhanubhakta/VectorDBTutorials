```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

# Define two vectors
vec1 = [2, 2]
vec2 = [3, 5]

# Function to calculate cosine similarity manually
def cosine_similarity(v1, v2):
    dot_product = sum(a * b for a, b in zip(v1, v2))
    norm_v1 = sqrt(sum(a * a for a in v1))  # Manual calculation of norm
    norm_v2 = sqrt(sum(b * b for b in v2))  # Manual calculation of norm
    cos_sim = dot_product / (norm_v1 * norm_v2)
    return cos_sim

# Function to calculate Euclidean distance manually
def euclidean_distance(v1, v2):
    diff = [a - b for a, b in zip(v1, v2)]
    squared_diff = [x * x for x in diff]
    sum_squared_diff = sum(squared_diff)
    euclid_dist = sqrt(sum_squared_diff)
    return euclid_dist

# Function to calculate Manhattan distance without numpy
def manhattan_distance(v1, v2):
    return sum(abs(a - b) for a, b in zip(v1, v2))

# Calculate metrics
cos_sim = cosine_similarity(vec1, vec2)
euclid_dist = euclidean_distance(vec1, vec2)
manhat_dist = manhattan_distance(vec1, vec2)

# Plot vectors
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, vec1[0], vec1[1], angles='xy', scale_units='xy', scale=1, color='r', label=f'Vector 1: {vec1}')
plt.quiver(0, 0, vec2[0], vec2[1], angles='xy', scale_units='xy', scale=1, color='b', label=f'Vector 2: {vec2}')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.legend()
plt.title('Vector Plot with Cosine Similarity, Euclidean, and Manhattan Distance')

# Annotate metrics
plt.annotate(f'Cosine Similarity: {cos_sim:.2f}', (4.5, 3.5), textcoords="offset points", xytext=(0,10), ha='center')
plt.annotate(f'Euclidean Distance: {euclid_dist:.2f}', (4.5, 3.0), textcoords="offset points", xytext=(0,10), ha='center')
plt.annotate(f'Manhattan Distance: {manhat_dist:.2f}', (4.5, 2.5), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```
